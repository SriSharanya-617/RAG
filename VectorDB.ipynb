```python
chunks = [

    "Employees receive 12 casual leaves annually.",

    "Employees receive 15 sick leaves annually.",

    "Employees may work from home twice per week.",

    "Travel expenses are reimbursed within 30 days.",

    "All employees are covered under company medical insurance."

]

print("Total Chunks:", len(chunks))
```

Load Embedding Model

```python
from sentence_transformers import SentenceTransformer

model = SentenceTransformer(
"all-MiniLM-L6-V2"
)
print("Embedding Model Loaded")
```

Generate Embeddings

```python
embeddings=model.encode(chunks)
print("Embedding SHape:")
print(embeddings. shape)
```

 Create FAISS Index

```python
!pip install faiss-cpu
```

```python
import faiss

dimension=embeddings.shape[1]

index=faiss.IndexFlatL2(dimension)

print("FAISS Index Created")
```

Add Embeddings to FAISS

```python
import numpy as np
index. add(
np.array(embeddings,
dtype=np.float32)

)

print("vectors Stored:",index.ntotal)
```

user query

```python
query="How many casual leaves are allowd?"
print(query)
```

convert query into embeddings

```python
query_embedding=model.encode(
[query]
)
```

search faiss

```python
k=3
distances, indices=index.search(
np.array(
query_embedding,
dtype=np.float32
),
k
)
```

display results

```python
for i in range(k):
    print(f"Rank {i+1}:")
    print("Distance:", distances[0][i])
    print("Chunk:", chunks[indices[0][i]])
    print("-" * 50)
```
